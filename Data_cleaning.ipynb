## Let's See The Visualization of Null Values

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.show()
```

### See We can not replace catagorical data, with some random assumption,in case of name of director or country. It is better for us to remove those rows.

```python
df.dropna(inplace=True)
```

### Convert Date to dd/mm/yyyy Format

```python
df["date_added"] = df["date_added"].str.strip()
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")
df["date_added"] = df["date_added"].dt.strftime("%d/%m/%Y")
```

```python
df["date_added"] = pd.to_datetime(df["date_added"], dayfirst=True)
```
